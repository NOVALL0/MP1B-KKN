```python
import pandas as pd


file_path = '/content/drive/MyDrive/asthma_dataset.csv'

# Memuat data CSV ke DataFrame
df = pd.read_csv(file_path)

# Menampilkan 5 baris pertama untuk melihat data
df.head()
```

```python
# Menghapus baris yang mengandung NaN
df_cleaned = df.dropna()

# Menampilkan data yang sudah dibersihkan
df_cleaned.head()
```

```python
# Impor seaborn dan matplotlib untuk visualisasi
import seaborn as sns
import matplotlib.pyplot as plt

# Tentukan ukuran grafik
plt.figure(figsize=(10, 6))

# Buat scatter plot dengan variabel 'Age', 'Peak_Flow', dan 'Asthma_Diagnosis'
sns.scatterplot(x='Age', y='Peak_Flow', hue='Asthma_Diagnosis', data=df, palette='Set2', style='Asthma_Diagnosis', s=40)

# Tambahkan legenda dengan judul 'Asma' di pojok kanan atas
plt.legend(title='Asma', loc='upper right')

# Tambahkan judul dan label sumbu
plt.title('Sebaran Data Age, Peak Flow, dan Asma', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Peak Flow', fontsize=12)

# Tampilkan grid dan grafik
plt.grid(True)
plt.show()
```

```python

# preprocessing
# Impor library numpy untuk manipulasi array
import numpy as np

# Memilih kolom 'Age' dan 'Peak_Flow' dari DataFrame dan mengonversinya menjadi array numpy untuk fitur (x_train)
x_train = np.array(df[['Age','Peak_Flow']])

# Memilih kolom 'Asthma_Diagnosis' dari DataFrame dan mengonversinya menjadi array numpy untuk label (y_train)
y_train = np.array(df['Asthma_Diagnosis'])

# Mencetak nilai x_train yang berisi data fitur
print(f'x_train:\n{x_train}\n')

# Mencetak nilai y_train yang berisi data label (diagnosis asma)
print(f'y_train:\n{y_train}\n')


```

```python
# Impor LabelBinarizer dari sklearn untuk mengubah label kategori menjadi format biner
from sklearn.preprocessing import LabelBinarizer

# Membuat objek LabelBinarizer untuk mengubah label menjadi biner
lb = LabelBinarizer()

# Mengubah y_train yang awalnya berupa label kategori menjadi format biner
y_train = lb.fit_transform(y_train)

# Mencetak hasil y_train yang sudah dibinarisasi
print(f'y_train:\n{y_train}\n')
```

```python
# Mengubah bentuk (reshape) y_train menjadi array 1 dimensi menggunakan flatten()
y_train = y_train.flatten()

# Mencetak hasil y_train yang sudah dalam bentuk 1 dimensi
print(f'y_train: {y_train}')
```

```python
#training kkn
from sklearn.neighbors import KNeighborsClassifier
# Impor KNeighborsClassifier dari sklearn untuk membuat model K-Nearest Neighbors (KNN)
from sklearn.neighbors import KNeighborsClassifier

# Menentukan jumlah tetangga terdekat (K) yang digunakan dalam model KNN
K = 3

# Membuat objek model KNN dengan jumlah tetangga terdekat K
model = KNeighborsClassifier(n_neighbors=K)

# Melatih model menggunakan data pelatihan (x_train untuk fitur dan y_train untuk label)
model.fit(x_train, y_train)
```

```python
# prediksi penderita asma
# Mendefinisikan nilai baru untuk prediksi berdasarkan umur (Age) dan nilai peak flow
Age = 30
Peak_Flow = 100

# Membentuk array 2D dengan data baru, sehingga sesuai dengan format input model
X_new = np.array([Age, Peak_Flow]).reshape(1, -1)

# Mencetak data baru yang akan diprediksi
X_new

```

```python
# Menggunakan model yang sudah dilatih untuk memprediksi label (diagnosis asma) berdasarkan data baru (X_new)
y_new = model.predict(X_new)

# Mencetak hasil prediksi untuk data baru
y_new
```

```python
# Mengonversi hasil prediksi yang berupa nilai biner (0 atau 1) kembali ke label kategori menggunakan LabelBinarizer
lb.inverse_transform(y_new)
```

```python
# Impor matplotlib untuk membuat plot
import matplotlib.pyplot as plt

# Membuat objek figure dan axis untuk plot
fig, ax = plt.subplots()

# Membuat scatter plot dari data training (x_train), dengan warna berdasarkan y_train
# x_train[:, 0] adalah 'Age' dan x_train[:, 1] adalah 'Peak_Flow'
ax.scatter(x_train[:, 0], x_train[:, 1], c=y_train, cmap='Set2', edgecolor='k')

# Menambahkan titik "Misterius" pada koordinat (35, 160) dengan marker kotak merah
ax.scatter(35, 160, marker='s', color='red', label='Misterius')

# Menambahkan legenda di bagian kiri atas dengan judul 'Keterangan'
ax.legend(loc='upper left', title='Keterangan')

# Memberikan judul pada plot
ax.set_title('Sebaran Data Age, Peak Flow, dan Diagnosis Asma', fontsize=14)

# Memberikan label pada sumbu x
ax.set_xlabel('Age', fontsize=12)

# Memberikan label pada sumbu y
ax.set_ylabel('Peak Flow', fontsize=12)

# Menampilkan grid untuk memudahkan pembacaan data
ax.grid(True)

# Menampilkan plot
plt.show()
```

```python
# Menambahkan kolom baru 'jarak' ke DataFrame df yang berisi jarak Euclidean antara 'misterius' dan setiap titik data
df['jarak'] = data_jarak

# Mengurutkan DataFrame berdasarkan kolom 'jarak' secara default dari yang terkecil
df_sorted = df.sort_values('jarak')

# Mencetak DataFrame yang sudah diurutkan berdasarkan jarak
df_sorted
```

evaluasi KKN clasification model

```python
# testing set
# Membuat array x_test yang berisi data fitur baru (Age dan Peak Flow) untuk testing
x_test =  np.array([[30,160],[40,168],[54,200],[34,90]])

# Mengonversi label kategori (misalnya 'Tidak Asma' dan 'Asma') menjadi format biner menggunakan LabelBinarizer
y_test = lb.transform(np.array(['Tidak Asma','Tidak Asma','Asma','Tidak Asma'])).flatten()

# Mencetak data fitur x_test
print(f'x_test:\n{x_test}\n')

# Mencetak label y_test yang sudah dibinarisasi
print(f'y_test: {y_test}')

```

```python
# prediksi testing set
y_pred = model.predict(x_test)
y_pred
```

```python
# Acuracy

# Mengimpor fungsi accuracy_score dari sklearn.metrics untuk menghitung akurasi model
from sklearn.metrics import accuracy_score

# Menghitung akurasi model dengan membandingkan label prediksi (y_pred) dan label sebenarnya (y_test)
acc = accuracy_score(y_test, y_pred)

# Mencetak hasil akurasi
acc
```

```python
# precission
# Mengimpor fungsi precision_score dari sklearn.metrics untuk menghitung presisi model
from sklearn.metrics import precision_score

# Menghitung presisi model dengan membandingkan label sebenarnya (y_test) dan label prediksi (y_pred)
prec = precision_score(y_test, y_pred)

# Mencetak nilai presisi
print(f'precision: {prec}')
```

```python
# recall
# Mengimpor fungsi recall_score dari sklearn.metrics untuk menghitung recall model
from sklearn.metrics import recall_score

# Menghitung recall model dengan membandingkan label sebenarnya (y_test) dan label prediksi (y_pred)
recall = recall_score(y_test, y_pred)

# Mencetak nilai recall
print(f'recall: {recall}')
```

```python
# Mengimpor fungsi f1_score dari sklearn.metrics untuk menghitung F1-Score model
from sklearn.metrics import f1_score

# Menghitung F1-Score model dengan membandingkan label sebenarnya (y_test) dan label prediksi (y_pred)
f1 = f1_score(y_test, y_pred)

# Mencetak nilai F1-Score
print(f'f1 score: {f1}')
```

```python
# Mengimpor fungsi classification_report dari sklearn.metrics untuk mendapatkan laporan klasifikasi
from sklearn.metrics import classification_report

# Menghasilkan laporan klasifikasi yang mencakup precision, recall, F1-score, dan support
cr = classification_report(y_test, y_pred)

# Mencetak laporan klasifikasi
print(f'classification report:\n{cr}')
```

```python
# Mengimpor fungsi matthews_corrcoef dari sklearn.metrics untuk menghitung Matthews Correlation Coefficient (MCC)
from sklearn.metrics import matthews_corrcoef

# Menghitung MCC antara label sebenarnya (y_test) dan label prediksi (y_pred)
mcc = matthews_corrcoef(y_test, y_pred)

# Mencetak nilai MCC
print(f'mcc: {mcc}')
```
